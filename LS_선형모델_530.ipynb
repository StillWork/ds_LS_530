```python
x1 = np.random.rand(5)
x2 = np.random.rand(5)
y = 2*x1 + 3*x2 + 4
x1,x2,y
```

```python
# shift-M 셀 합치기
# shist-ctrl-'-'  셀 나누기
# a , b  셀 추가
# x 삭제,  o 출력보기

w1 = np.random.rand()
w2 = np.random.rand()
b = np.random.rand()

n_epoch = 20
lr = 0.1
for epoch in range(n_epoch):
    y_pred = w1*x1 + w2*x2 + b
    e = (y_pred - y)
    w1 = w1 - lr*(e*x1).mean()
    w2 = w2 - lr*(e*x2).mean()
    b = b - lr*e.mean()

    error = np.abs(e).mean()
    print('error', error, 'w1=', w1, 'w2=', w2, 'b=', b)
```

```python
x1, x2
```

```python
X = np.vstack((x1,x2)).T
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_, model.intercept_
```

```python
model.predict(X)
```
